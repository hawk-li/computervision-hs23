# Lecture 03: Image Geometry

This notebook features a quick example on how you can perform geometric transformations. This example is based on: https://scikit-image.org/docs/stable/auto_examples/transform/plot_geometric.html#sphx-glr-auto-examples-transform-plot-geometric-py

```python
from skimage import io
from skimage import transform
from skimage import img_as_float

import numpy as np
import matplotlib.pyplot as plt

img = io.imread('data/IMG_20230622_085147.jpg')

img.shape
```

```python
plt.imshow(img)
```

We define arrays that contain image locations that we would like to warp:

```python
src = np.array([[0, 0], [400, 0], [400, 100], [0, 100]])
dst = np.array([[500, 1800], [2400, 2000], [2400, 1200], [500, 1150]])
```

```python
f, ax = plt.subplots(1, 1, figsize=(20, 10))
ax.imshow(img)
ax.scatter(dst[:, 0], dst[:, 1], color='red')
```

We warp the facade of the building into a rectangular projection, as if we would stand right in front of it.

```python

tform = transform.ProjectiveTransform()
tform.estimate(src, dst)
warped = transform.warp(img, tform, output_shape=(100, 400))
```

```python
f, ax = plt.subplots(1, 1, figsize=(20, 4))
ax.imshow(warped[::-1])
```
